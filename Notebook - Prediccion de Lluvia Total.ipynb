b) cross_val_score:
Ejecute el siguiente tuning de parámetros (learning curves - notebook 10, "https://gitlab.com/bivl2ab/academico/cursos-uis/ai/ai-uis-student/-/blob/master/10_std_Notes_Regression.ipynb") para determinar el MEJOR número de folds para cada estimador, con parámetros por defecto:
   - DecisionTree
   - RandomForest
   - SVC/R

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

# ... (código previo para cargar datos y definir X, y) ...

def evaluate_folds(estimator, X, y, folds_range):
  """Evalúa un estimador con diferentes números de folds."""
  results = []
  for n_folds in folds_range:
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)
    scores = cross_val_score(estimator, X, y, cv=kf, scoring='neg_mean_squared_error')
    mean_score = -scores.mean()  # Negamos para obtener MSE positivo
    results.append((n_folds, mean_score))
  return results

# Rango de folds a evaluar
folds_range = range(2, 11)

# 1. DecisionTreeRegressor
tree_results = evaluate_folds(DecisionTreeRegressor(), X, y, folds_range)
best_tree_folds = min(tree_results, key=lambda item: item[1])[0]
print(f"Mejor número de folds para DecisionTree: {best_tree_folds}")

# 2. RandomForestRegressor
forest_results = evaluate_folds(RandomForestRegressor(), X, y, folds_range)
best_forest_folds = min(forest_results, key=lambda item: item[1])[0]
print(f"Mejor número de folds para RandomForest: {best_forest_folds}")

# 3. SVR
svr_results = evaluate_folds(SVR(), X, y, folds_range)
best_svr_folds = min(svr_results, key=lambda item: item[1])[0]
print(f"Mejor número de folds para SVR: {best_svr_folds}")
```

# Algoritmo genetico


```python
# Importar librerías necesarias
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import random

# Configuración de gráficas
plt.style.use('ggplot')

# Cargar el dataset meteorológico (reemplaza con la ruta de tu archivo)
data = meteor

# Mostrar las primeras filas para ver la estructura
print(data.head())

# Preprocesamiento
X = data.drop("Total Rainfall(mm)", axis=1)  # Excluir la variable dependiente (en este caso "Lluvia")
y = data["Total Rainfall(mm)"]  # Variable dependiente (cantidad de lluvia)

# Estandarización de los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Función para Reducción de Dimensionalidad (PCA)
def apply_pca(data, n_components):
    pca = PCA(n_components=n_components)
    data_reduced = pca.fit_transform(data)
    return data_reduced

# Función para Regresión con DecisionTreeRegressor
def regression_model(X_train, y_train, X_test, y_test, n_estimators):
    model = DecisionTreeRegressor(max_depth=n_estimators, random_state=42)
    model.fit(X_train, y_train)

    pca = PCA(n_components=X_train.shape[1])
    X_test_reduced = pca.fit_transform(X_test)

    predictions = model.predict(X_test_reduced)
    mse = mean_squared_error(y_test, predictions)
    return mse

# Algoritmo Genético para optimización de parámetros
def genetic_algorithm(X, y, population_size, generations, top_k):
    history_scores = []
    best_scores = []
    mean_scores = []

    # Inicializar la población
    population = [
        {"n_components": random.randint(2, min(X.shape[1], 10)),
         "n_estimators": random.randint(2, 20)}
        for _ in range(population_size)
    ]

    # Dividir los datos en entrenamiento y prueba
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    for generation in range(generations):
        results = []
        generation_scores = []

        for chromosome in population:
            n_components = chromosome["n_components"]
            n_estimators = chromosome["n_estimators"]

            # Aplicar PCA
            X_reduced = apply_pca(X_train, n_components)

            # Entrenar el modelo de regresión y obtener el error cuadrático medio
            mse = regression_model(X_reduced, y_train, X_test, y_test, n_estimators)
            results.append((chromosome, mse))
            generation_scores.append(mse)

        # Ordenar resultados por el MSE (menor es mejor)
        results.sort(key=lambda x: x[1])
        best_chromosomes = results[:top_k]

        # Guardar los mejores scores
        best_score = best_chromosomes[0][1]
        history_scores.append(best_score)
        best_scores.append(best_score)
        mean_scores.append(np.mean(generation_scores))

        # Reproducción (selección de los mejores)
        next_generation = [result[0] for result in best_chromosomes]
        while len(next_generation) < population_size:
            parent1, parent2 = random.sample(next_generation, 2)
            child = {
                "n_components": random.choice([parent1["n_components"], parent2["n_components"]]),
                "n_estimators": random.choice([parent1["n_estimators"], parent2["n_estimators"]])
            }
            if random.random() < 0.2:  # Mutación
                child["n_components"] = random.randint(2, min(X.shape[1], 10))
            if random.random() < 0.2:
                child["n_estimators"] = random.randint(2, 20)
            next_generation.append(child)

        population = next_generation

        print(f"Generación {generation + 1} - Mejor MSE: {best_score:.4f}")

    return best_chromosomes[0][0], history_scores, best_scores, mean_scores, X_train, X_test, y_train, y_test

# Ejecutar el Algoritmo Genético
best_solution, history_scores, best_scores, mean_scores, X_train, X_test, y_train, y_test = genetic_algorithm(
    X_scaled, y, population_size=20, generations=10, top_k=5
)

# Función para graficar la evolución de los scores
def plot_evolution(scores, best_scores, mean_scores):
    plt.figure(figsize=(10, 6))

    plt.plot(scores, 'o-', label="Mejor MSE", color="blue")
    plt.xlabel("Generación")
    plt.ylabel("Mejor MSE")
    plt.title("Evolución del Mejor MSE por Generación")
    plt.legend()
    plt.show()

    # Gráfico de los mejores scores y el promedio
    plt.figure(figsize=(10, 6))
    plt.plot(best_scores, label="Mejor Score", color='blue', marker='o')
    plt.plot(mean_scores, label="Promedio de Scores", color='orange', marker='x')
    plt.xlabel("Generación")
    plt.ylabel("MSE")
    plt.title("Mejor MSE y Promedio de MSE por Generación")
    plt.legend()
    plt.show()

# Graficar la evolución del MSE
plot_evolution(history_scores, best_scores, mean_scores)

# Usar el mejor modelo encontrado para hacer predicciones
n_components_best = best_solution["n_components"]
n_estimators_best = best_solution["n_estimators"]

# Aplicar PCA y entrenar el modelo con los mejores parámetros
X_reduced_best = apply_pca(X_scaled, n_components_best)
final_model = DecisionTreeRegressor(max_depth=n_estimators_best, random_state=42)
final_model.fit(X_reduced_best, y)

# Hacer predicciones sobre el conjunto de prueba
X_test_reduced = apply_pca(X_test, n_components_best)
predictions = final_model.predict(X_test_reduced)

# Mostrar las predicciones y calcular el error final
mse_final = mean_squared_error(y_test, predictions)
print(f"Error cuadrático medio final: {mse_final:.4f}")
```
